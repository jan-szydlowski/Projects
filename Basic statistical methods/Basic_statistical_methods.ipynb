```python
import numpy as np
from scipy.stats import norm, expon, chi2, uniform
import matplotlib.pyplot as plt
%matplotlib inline
```

# Standard distributions sampling


```python
#Normal distribution

# Create a normal distribution with mean = 1 and std dev = 4
rv_norm = norm(loc = 1, scale = 4)
x_norm = np.linspace(rv_norm.ppf(0.01), rv_norm.ppf(0.99), 100)
fig, ax = plt.subplots(1, 1)

# Plot the theoretical Normal Probability Density Function (PDF) as a red line
ax.plot(x_norm, rv_norm.pdf(x_norm), 'r-', lw=5, alpha=0.6, label='norm pdf')

r_norm = rv_norm.rvs(size=1000)
ax.hist(r_norm, density=True, histtype='stepfilled', alpha=0.2, bins = x_norm[::4])


ax.legend(loc='best', frameon=False)
plt.show()

# Comparing theoretical statistics vs. actual sample statistics
print("Distribution mean: {0}".format(rv_norm.stats(moments='m')))
print("Distribution variance: {0}".format(rv_norm.stats(moments='v')))
print("Sample mean: {0}".format(np.mean(r_norm)))
print("Sample variance: {0}".format(np.var(r_norm)))
```

```python
#Exponential
rv_exp=expon(scale=4) # Create a exponential random variable with scale 4
x_exp=np.linspace(rv_exp.ppf(0.01), rv_exp.ppf(0.99), 100) #Drawing function between 0.01 and 0.99 quantilesand with 100 points
fig, ax=plt.subplots(1,1)
ax.plot(x_exp, rv_exp.pdf(x_exp), 'r-', lw=5, alpha=0.6, label='expon pdf')

r_exp=rv_exp.rvs(size=1000)

# Plot the theoretical Probability Density Function (PDF) as a red line
ax.hist(r_exp, density=True, histtype='stepfilled', alpha=0.2, bins = x_exp[::4])

ax.legend(loc='best', frameon=False)
plt.show()

# Comparing theoretical statistics vs. actual sample statistics
print("Distribution mean: {0}".format(rv_exp.stats(moments='m')))
print("Distribution variance: {0}".format(rv_exp.stats(moments='v')))
print("Sample mean: {0}".format(np.mean(r_exp)))
print("Sample variance: {0}".format(np.var(r_exp)))
```

```python
#chi2
rv_chi2=chi2(df=8) #creating random variable chi2 function with 8 degrees of freedom
x_chi2=np.linspace(rv_chi2.ppf(0.01), rv_chi2.ppf(0.99), 100) #Drawing function between 0.01 and 0.99 quantiles and with 100 points
fig, ax=plt.subplots(1,1)

# Plot the theoretical Normal Probability Density Function (PDF) as a red line
ax.plot(x_chi2, rv_chi2.pdf(x_chi2), 'r-', lw=5, alpha=0.6, label='chi2 pdf')

r_chi2=rv_chi2.rvs(size=1000)
ax.hist(r_chi2, density=True, histtype='stepfilled', alpha=0.2, bins = x_chi2[::4])

ax.legend(loc='best', frameon=False)
plt.show()

# Comparing theoretical statistics vs. actual sample statistics
print("Distribution mean: {0}".format(rv_chi2.stats(moments='m')))
print("Distribution variance: {0}".format(rv_chi2.stats(moments='v')))
print("Sample mean: {0}".format(np.mean(r_chi2)))
print("Sample variance: {0}".format(np.var(r_chi2)))
```

```python
#uniform
rv_uni=uniform(loc=(-2*np.sqrt(3)), scale=(8*np.sqrt(3))) #creating random variable uniform function
x_uni=np.linspace(rv_uni.ppf(0.01), rv_uni.ppf(0.99), 200) #Drawing function between 0.01 and 0.99 quantiles and with 100 points
fig, ax=plt.subplots(1,1)

# Plot the theoretical Normal Probability Density Function (PDF) as a red line
ax.plot(x_uni, rv_uni.pdf(x_uni), 'r-', lw=5, alpha=0.6, label='uni pdf')

r_uni=rv_uni.rvs(size=1000)
ax.hist(r_uni, density=True, histtype='stepfilled', alpha=0.2, bins = x_uni[::4])

ax.legend(loc='best', frameon=False)
plt.show()

# Comparing theoretical statistics vs. actual sample statistics
print("Distribution mean: {0}".format(rv_uni.stats(moments='m')))
print("Distribution variance: {0}".format(rv_uni.stats(moments='v')))
print("Sample mean: {0}".format(np.mean(r_uni)))
print("Sample variance: {0}".format(np.var(r_uni)))
```

# Law of large numbers. Central Limit Theorem.

```python
#Drawing N samples from previously defined distributions. Ploting sample variance and mean against the size of the sample.

var_norm, var_exp, var_chi2, var_uni=[] ,[], [], []
mean_norm, mean_exp, mean_chi2, mean_uni=[], [], [], []
Ns=np.arange(2, 1001)
for N in range(2, 1001):
  #Samples for normal dist
  sample_size_norm=rv_norm.rvs(size=N)
  var_norm.append(np.var(sample_size_norm))
  mean_norm.append(np.mean(sample_size_norm))

  #samples for exponental dist
  sample_size_exp=rv_exp.rvs(size=N)
  var_exp.append(np.var(sample_size_exp))
  mean_exp.append(np.mean(sample_size_exp))

  #samples for chi2 dist
  sample_size_chi2=rv_chi2.rvs(size=N)
  var_chi2.append(np.var(sample_size_chi2))
  mean_chi2.append(np.mean(sample_size_chi2))

  #samples for unimodal dist
  sample_size_uni=rv_uni.rvs(size=N)
  var_uni.append(np.var(sample_size_uni))
  mean_uni.append(np.mean(sample_size_uni))

#Making plots
fig1, axs1 = plt.subplots(2, 2, figsize=(8, 6), dpi=100)

plots = [
    (0, 0, var_norm, "Normal"),
    (0, 1, var_exp,  "Exponential"),
    (1, 0, var_chi2, "Chi2"),
    (1, 1, var_uni,  "Uniform"),
]

for r, c, y, title in plots:
    ax = axs1[r, c]
    ax.plot(Ns, y)
    ax.set_xlabel("Sample size")
    ax.set_ylabel("Variance")
    ax.set_title(title)
    ax.grid(True)

fig.suptitle("Variances")
fig.tight_layout()
plt.show()

fig2, axs2 = plt.subplots(2, 2, figsize=(8, 6), dpi=100)
plots = [
    (0, 0, mean_norm, "Normal"),
    (0, 1, mean_exp,  "Exponential"),
    (1, 0, mean_chi2, "Chi2"),
    (1, 1, mean_uni,  "Uniform"),
]
for r, c, y, title in plots:
    ax = axs2[r, c]
    ax.plot(Ns, y)
    ax.set_xlabel("Sample size")
    ax.set_ylabel("Mean")
    ax.set_title(title)
    ax.grid(True)

fig.suptitle("Means")
fig.tight_layout()
plt.show()
```

```python
#Implementing function accept_rejection_sampling which samples from any bounded distribution on the [0,1] interval.
# Defining the target probability density function (PDF)
def toy_dist(x):

    # Normalization constant C to ensure the area under the curve equals 1
    C = 1/3 - np.exp(-4)/4 + 1/4
    return (x**2 + np.exp(-4*x)) / C

x = np.linspace(0,1,1000)

def accept_rejection_sampling(dist, size = 1):
    # Finding the maximum value of the distribution (M) to define the bounding box height
    M = np.max(dist)

    # Defining a proposal distribution (uniform across the [0, 1] interval)
    t_uni = uniform(loc=0, scale=1)

    # Defining a vertical uniform distribution to pick a random height up to M
    u_uni = uniform(loc=0, scale=M)

    solution = []

    # Keep sampling until we reach the desired number of accepted values
    while len(solution) < size:
        # 1. Generate a candidate value from the proposal distribution (X-axis)
        sample_t = t_uni.rvs(size=1)[0]

        # 2. Generate a random height for this candidate (Y-axis)
        sample_u = u_uni.rvs(size=1)[0]

        # 3. Acceptance condition: if the random height is below the target curve, accept it
        if toy_dist(sample_t) > sample_u:
            solution.append(sample_t)

    return solution



samples = accept_rejection_sampling(toy_dist(x), size = 10000)

fig, ax = plt.subplots(1, 1)
ax.plot(x, toy_dist(x), 'r-', lw=5, alpha=0.6, label='toy pdf')
ax.hist(samples, density=True, histtype='stepfilled', alpha=0.2, bins = x[::40], label = "sampled")
ax.legend(loc='best', frameon=False)
plt.show()
```

```python
# Implementing Monte-Carlo procedure for solving 2D Laplace equation with Dirichlet boundary conditions

# Initialize the grid to store temperature/potential values (plane)
# and a grid to count visits or hits for averaging (plane_counts)
S = 100 # Decrease if computations take too long
plane = np.zeros((S,S))
plane_counts = np.zeros((S,S))

# Set Dirichlet boundary conditions (fixed values on the edges of the square):
plane[0] = np.cos(np.linspace(-1.5,1.5,S)*2*np.pi)
plane[-1] = -np.cos(np.linspace(-1,1,S)*2*np.pi)
plane[:,0] = np.cos(np.linspace(-0.5,0.5,S)*2*np.pi)
plane[:,-1] = -2+np.linspace(-1,1,S)**2
```

```python
plt.figure(figsize=(10,15))
plt.title("0 walks")
plt.imshow(plane)
```

```python

# Random Generators

start_gen = uniform(loc=0, scale=S)

direction_gen = uniform(loc=0, scale=100)

def monte_carlo(number):
    for i in range(number):
        # 1. Select a random starting point (x, y) inside the grid
        x_start = int(start_gen.rvs(size=1)[0])
        y_start = int(start_gen.rvs(size=1)[0])

        current_point = [x_start, y_start]
        traveled_way = []

        # 2. Random Walk Loop: Move until the point hits any of the four boundaries
        while 0 < current_point[0] < S-1 and 0 < current_point[1] < S-1:

            # Record each visited position to update it later
            traveled_way.append(list(current_point))

            # Generate a random number to choose a direction
            rand_val = direction_gen.rvs(size=1)[0]

            if rand_val > 75:
                current_point[1] += 1       # Up
            elif rand_val > 50:
                current_point[1] -= 1       # Down
            elif rand_val > 25:
                current_point[0] += 1       # Right
            else:
                current_point[0] -= 1       # Left

        # Hitting wall
        x_end, y_end = int(current_point[0]), int(current_point[1])
        v_b = plane[x_end, y_end]

        # 4. Back-propagation: Update all points visited during the walk
        for (px, py) in traveled_way:
            px, py = int(px), int(py)

            v_old = plane[px, py]
            n = plane_counts[px, py]

            # Update the value using an incremental average (running mean)
            # Each point becomes an average of the boundary values its walks eventually hit
            plane[px, py] = (n * v_old + v_b) / (n + 1)
            plane_counts[px, py] += 1

# Plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
#1 walk
axes[0].imshow(plane, origin='upper', interpolation='nearest')
axes[0].set_title("1 walks")

monte_carlo(5) #5 walks
axes[1].imshow(plane, origin='upper', interpolation='nearest')
axes[1].set_title("5 walks")

monte_carlo(33) #33 walks
axes[2].imshow(plane, origin='upper', interpolation='nearest')
axes[2].set_title("33 walks")

plt.show()
```
