```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("kc_house_data.csv") #Reading file
data.head(6)
```

```python
print(data.columns) #Assigning data to categorical or numerical data
house_categorical_df = data[['id', 'date', 'waterfront', 'zipcode', 'view', 'condition', 'grade']]
house_numerical_df = data[['bathrooms', 'bedrooms', 'floors', 'lat', 'long', 'price', 'sqft_above', 'sqft_basement', 'sqft_living', 'sqft_living15', 'sqft_lot', 'sqft_lot15', 'yr_built', 'yr_renovated']]
```

```python
house_numerical_df.describe()
#General statistics of df
```

```python
relation = data.plot.scatter(y='price', x='sqft_living')
relation.ticklabel_format(style='plain', axis='y', useOffset=False)
#Showing relation between square footage and price
```

```python
#Finging other variables correlated with price
price_and_grade = data.plot.scatter(y='price', x='grade')
price_and_grade.ticklabel_format(style='plain', axis='y', useOffset=False)

price_and_bathrooms = data.plot.scatter(y='price', x='bathrooms')
price_and_bathrooms.ticklabel_format(style='plain', axis='y', useOffset=False)

price_and_sqft_lot = data.plot.scatter(y='price', x='sqft_lot')
price_and_sqft_lot.ticklabel_format(style='plain', axis='y', useOffset=False)
```

```python
#Checking how good apartments are distributed over the city
#By zip code:
price_and_zipcode = data.plot.scatter(y='price', x='zipcode')
price_and_zipcode.ticklabel_format(style='plain', axis='y', useOffset=False)
```

```python
#By lattitude and longtitude
latt_and_long = data.plot.scatter(y='lat', x='long', alpha=0.1, color ='green')
latt_and_long.ticklabel_format(style='plain', axis='y', useOffset=False)
```

```python
#Creating boxplot of apartment's grade versus it's price
price_grade_boxplt=data.boxplot(column='price', by='grade', grid=False)
price_grade_boxplt.ticklabel_format(style='plain', axis='y', useOffset=False)

price_and_grade = data.plot.scatter(y='price', x='grade')
price_and_grade.ticklabel_format(style='plain', axis='y', useOffset=False)
```

```python
from statsmodels.graphics.mosaicplot import mosaic

#Creating mosaic plot
mosaic(data, ['view', 'condition'])
plt.show()
```

```python
#Creating a condition for specific house (3 bedrooms, 2 bathrooms and 2 floors)
fulfiling_aparts = data.loc[(data['bedrooms']==3) & (data['bathrooms']==2) & (data['floors']==2)]
fulfiling_aparts.describe()
```

```python
# Distinct accepted apartments
pg_condition  = (data['bedrooms']==3) & (data['bathrooms']==2) & (data['floors']==2)
subset = data[pg_condition]

basic_plot = data.plot.scatter(x='grade', y='price')
subset.plot.scatter(x='grade', y='price', color='yellow', ax=basic_plot) # ax=basic_plot makes that subset plot is drawn on basic_plot plot
```

```python
#Finding apartments with best price to footage ratio
price_to_footage = (subset
                     .assign(best_ratio= subset['price']/subset['sqft_living'])
                     .sort_values(by=['best_ratio'], ascending=[True]))
price_to_footage.head()
```
